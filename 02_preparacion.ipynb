```python
import pandas as pd
```

# Cargar datos del club

```python
datos_aura = pd.read_csv("partidos_aura.csv")
datos_aura.head()
```

# Revisar Información General

```python
datos_aura.info()
```

# Encontrar y manejar datos faltantes

```python
datos_aura.isnull().sum()
datos_aura['goles_local'] = datos_aura['goles_local'].fillna(datos_aura['goles_local'].mean())
print("Valores nulos antes de la imputación:")
print(datos_aura.isnull().sum())
```

# One-Hot Encoding para equipos

```python
datos_preparados = pd.get_dummies(datos_aura, columns=['equipo_local', 'equipo_visitante'], drop_first=True)
datos_preparados.head()
```

# Eliminar filas duplicadas

```python
datos_preparados.drop_duplicates(inplace=True)

print("Filas antes de eliminar duplicados:", len(datos_preparados))
datos_preparados.drop_duplicates(inplace=True)
print("Filas después de eliminar duplicados:", len(datos_preparados))
```

# Manejar Fechas

```python
datos_preparados['fecha_partido'] = pd.to_datetime(datos_preparados['fecha_partido'], errors='coerce')

print("Fechas invalidas (NaT) luego de la conversión:")
print(datos_preparados['fecha_partido'].isnull().sum())
```

# Resumen de la conversion
